### 1.2 首先定义几个辅助函数
#### 1.2.1 衡量聚类效果
#####   查阅文献找到对于数据集$D={x_1, x_2, ..., x_m}$, 假定通过聚类给出的划分为$C={C_1, C_2, ..., C_k}$, 参考模型 (即标记数据) 给出的划分为$C^*={C^*_1, C^*_2, ..., C^*_k}$. 相应得另$\lambda$与$\lambda^*$分别表示与$C$和$C^*$对应的簇标记向量, 下面定义:  
$$a=|SS|, SS={(x_i, x_j)|\lambda_i=\lambda_j, \lambda^*_i=\lambda^*_j, i<j}$$
$$b=|SD|, SD={(x_i, x_j)|\lambda_i=\lambda_j, \lambda^*_i \neq \lambda^*_j, i<j}$$
$$c=|DS|, DS={(x_i, x_j)|\lambda_i \neq \lambda_j, \lambda^*_i=\lambda^*_j, i<j}$$
$$d=|DD|, DD={(x_i, x_j)|\lambda_i \neq \lambda_j, \lambda^*_i \neq \lambda^*_j, i<j}$$
由于每个样本只能出现在一个集合中, 所以有:
$$a+b+c+d=\frac{m(m-1)}{2}$$
m为数据条数  
下面定义:
Jaccard Coefficient (JC):
    $$JC=\frac{a}{a+b+c}$$
FM Index (FMI):
    $$FMI=\sqrt{\frac{a}{a+b}\frac{a}{a+c}}$$
也可以看成
    $$FMI=\frac{TP}{\sqrt{(TP+FP)(TP+FN)}}$$
Rand Index (RI):
    $$RI=\frac{2(a+d)}{m(m-1)}$$
    
按照定义, 这三个数值都应该是越大越好

```python
def assess(predict_y, y, method='all'):
    '''
    @param:
        -predict_y: Should be the predicted data using clustering algorithm, and its dim should be m * 1
        -y: Original labeled data, and its dim also should be m * 1
        -method: Which metric will be used to measure the performance of the clustering algorithm
    '''
    m = predict_y.shape[0]
    a, b, c, d = 0, 0, 0, 0
    flag1, flag2 = None, None
    # calculate a, b, c and d
    for j in range(m):
        for i in range(j):
            flag1 = (predict_y[i, 0] == predict_y[j, 0])
            flag2 = (y[i, 0] == y[j, 0])
            if flag1 and flag2:
                a += 1
            if flag1 and (not flag2):
                b += 1
            if (not flag1) and flag2:
                c += 1
            if (not flag1) and (not flag2):
                d += 1
    if method == 'all':
        return [(a / (a + b + c)), (np.sqrt((a / (a + b)) * (a / (a + c)))), (2 * (a + d) / (a + b + c + d))]
    if method == 'JC':
        return a / (a + b + c)
    if method == 'FMI':
        return np.sqrt((a / (a + b)) * (a / (a + c)))
    if method == 'RI':
        return 2 * (a + d) / (a + b + c + d)
```

### 1.3 Mean Shift
##### Mean Shift算法是一种简单的聚类方法, 其中使用到了kernel trick用来计算将数据特征投影到高维后的内积相似度. 它每次以r=bandwidth为半径做圆形, 对区域内包含的点求中心, 作为下一次的圆心, 迭代求解

```python
from sklearn.cluster import MeanShift
# 改变bindwidth值, 即圆的半径来观察得到的结果
ys = []
scores = []
for i in np.arange(500, 3000, 500):
    meanshift = MeanShift(bandwidth=i, cluster_all=True)
    predict = meanshift.fit_predict(norm_x).reshape((-1, 1))
    ys.append(predict)
    scores.append(assess(predict, y))
scores = np.array(scores)
```

### 1.6 Affinity Propogation

##### AP算法的基本思想是将全部样本看作网络的节点，然后通过网络中各条边的消息传递计算出各样本的聚类中心。聚类过程中，共有两种消息在各节点间传递，分别是吸引度( responsibility)和归属度(availability) 。AP算法通过迭代过程不断更新每一个点的吸引度和归属度值，直到产生m个高质量的Exemplar（类似于质心），同时将其余的数据点分配到相应的聚类中。  

$$r(i, j)=s(i, j)-max\{a(i, j), s(i, j)\}, j\neq j'$$
$$ F^{HLLC}=\left\{
\begin{array}{rcl}
min\{0, r(j, j)+\sum_{i' \neq i,j}max\{0, r(i', j)\}\}       &      & {i\neq j}\\
\sum_{i' \neq i,j}max\{0, r(i', j)\}     &      & {i=j}
\end{array} \right. $$

$$R_{i+1}=(1-\lambda)R_i + \lambda R_{i-1}$$
$$A_{i+1}=(1-\lambda)A_i + \lambda A_{i-1}$$
##### $s{i, j}$是样本i和j的相似性度量, $\lambda \in [0, 1]$成为阻尼因子。$R=(r(i, j)), A=(a(i, j))$分别为吸引度矩阵和归属度矩阵, t代表迭代次数  
$$s(i, k)=-min||T_0 x_i - T x_k||^2$$
或
$$s(i, k)=m(i, k)-\frac{1}{N}\sum_{j=1}^N m(i, j)-\frac{1}{N}\sum_{j=1}^N m(j, k)$$
##### 其中,$T_0$和 $T$ 分别代表图像中心和图像的窗口参数; $m( i, k)$代表图像 i 和图像k满足特定匹配的特征个数

```python
import numpy as np
import pandas as pd
import xlrd
import matplotlib
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
matplotlib.rcParams['axes.unicode_minus']=False
%matplotlib inline
from sklearn.model_selection import train_test_split
```

### 2.1 数据读取与训练测试集分割

```python
def get_data():
    filepath = "../../有标签数据/"
    filename_np = "totalNPdata.xlsx"
    filename_pd = "totalPDSdat.xlsx"
    y = []
    np_file = xlrd.open_workbook(filepath+filename_np)
    pd_file = xlrd.open_workbook(filepath+filename_pd)
    table_np = np_file.sheets()[0]
    table_pd = pd_file.sheets()[0]
    nrows_np = table_np.nrows
    nrows_pd = table_pd.nrows
    ncols = table_np.ncols
    header = table_np.row_values(0)
    all_data_list = []
    for i in range(1,nrows_np):
        data = {}
        row_data = table_np.row_values(i)
        for i in range(1,ncols):
            data[header[i]] = row_data[i]
        all_data_list.append(data)
        y.append(0)
    for i in range(1,nrows_pd):
        data = {}
        row_data = table_pd.row_values(i)
        for i in range(1, ncols):
            data[header[i]] = row_data[i]
        all_data_list.append(data)
        y.append(1)
    c = list(zip(all_data_list, y))
    np.random.shuffle(c)
    all_data_list[:], y[:] =zip(*c)
    pd_data = pd.DataFrame(all_data_list)
    for h in header[1:]:
        pd_data[h] = (pd_data[h]-pd_data[h].min())/(pd_data[h].max()-pd_data[h].min())
    return pd_data, y
```

```python
data,y = get_data()
data_train, data_test, y_train, y_test = train_test_split(data, y,test_size=0.25)
```

#### 上面将数据按照0.75:0.25分割成训练集与测试集两个部分

### 2.2 下面进行kmeans聚类分析

#### 2.2.1 直接导入sklearn的kmeans算法包，进行聚类并取身高和步长两个维度进行可视化

```python
def kmeans_cluster(initial):
    kmeans = KMeans(n_clusters=2,init=initial)
    kmeans.fit(data_train)
    result = kmeans.predict(data_test)
    count_diff = 0
    count_same = 0
    for r, y_0 in zip(result, y_test):
        if r==y_0:
            count_same+=1
        else:
            count_diff+=1
    # print(count_same, count_diff)
    plt.subplot(221)
    plt.title("kmeans聚类结果")
    plt.scatter(data_test['身高'],data_test['步长（左）'],s= 30,c= result,linewidths=0)
    plt.subplot(222)
    plt.title("原数据的分布情况")
    plt.scatter(data_test['身高'],data_test['步长（左）'],s= 30,c= y_test,linewidths=0)
    plt.show()
    print("正确率：",count_same/(count_same+count_diff))
```

### 2.3 下面用SOM自组织映射进行聚类分析

#### 2.3.1 用训练数据进行训练

```python
#聚类应该聚为两类，则竞争层应该为两个结点
alpha = 0.001
#alpha为递减的学习速率
#输入层到竞争层的两个结点的连接权重分别为W_1，W_2
W_1 = np.random.normal(0.5, 0.4, size=[23])
W_2 = np.random.normal(0.5, 0.4, size=[23])
```

```python
def get_type_of_data(data_normal):
    #前向传播得到竞争胜利的结点 
    W_1_normal = W_1/np.sqrt(np.sum(np.square(W_1)))
    W_2_normal = W_2/np.sqrt(np.sum(np.square(W_2)))
    similarity_1 = np.sum(np.square(W_1_normal-data_normal))
    similarity_2 = np.sum(np.square(W_2_normal-data_normal))
    if similarity_1<similarity_2:
        return 0,W_1_normal,W_2_normal,similarity_1,similarity_2
    else:
        return 1,W_1_normal,W_2_normal,similarity_1,similarity_2
```

```python
def refresh(data):
    #更新竞争胜利结点的权重
    global W_1
    global W_2
    data_normal = data/np.sqrt(np.sum(np.square(data)))
    type_of_data, W_1_normal, W_2_normal,_,_ = get_type_of_data(data_normal)
    if type_of_data == 0:
        W_1 = W_1_normal + alpha * (data_normal -W_1_normal)
    else:
        W_2 = W_2_normal + alpha * (data_normal -W_2_normal)
```

```python
#对所有数据迭代训练200次
for step in range(200):
    for  index, d in data_train.iterrows():
        refresh(np.asarray(d))
    alpha = 0.5-(0.5-0.0001)*step/500
```

#### 2.3.2 用测试数据测试聚类的准确率

```python
def test():
    result = []
    for index, d in data_test.iterrows():
        i,_,_,s1,s2 =get_type_of_data(np.asarray(d)/np.sqrt(np.sum(np.square(d))))
        result.append(i)
    count_1 = 0
    count_2 = 0
    for r, y_0 in zip(result, y_test):
        if r==y_0:
            count_1+=1
        else:
            count_2+=1
    plt.subplot(223)
    plt.title("som聚类结果")
    plt.scatter(data_test['身高'],data_test['步长（左）'],s= 30,c= result,linewidths=0)
    plt.subplot(224)
    plt.title("原数据分布情况")
    plt.scatter(data_test['身高'],data_test['步长（左）'],s= 30,c= y_test,linewidths=0)
    plt.show()
    print("正确率：",count_2/(count_1+count_2))
```

```python
test()
```

#### SOM是一个网络模型，但是只有输入和竞争（输出）两层，在训练时需要进行，迭代调整权重。
#### 在实验中，当迭代次数为100时，正确率不到70%，当迭代次数为200时正确率，正确率到达77%，而进一步增加迭代次数，正确率几乎没有什么提升.
#### 试验中有时候会有权重初始化状态不好，导致最后聚类效果很差。

讨论K值对聚类的影响


从上面的结果中，可以看出当k=4、k=3时，聚类效果较好，因为族与族之间相对于其他k值，较为分散。同时，这也反应出kmeans的一个弱点：需要人为指定聚类数目K，如果k值指定的不好，聚类的效果也不是很好。其对初始中心的选择的依赖性较强，所以，产生局部的最优解发生的概率非常大。Kmeans对数据集合内的噪声和离群点的敏感较高，因为此类数据也许会对均值造成一定的影响
算法执行速度快时kmeans的一个显著优点，其时间复杂度为O（nkt），所以，在处理大型数据集时，它具有可伸缩性和高效性. 其中，n是样本的数目，k 是类簇的数目，t 是迭代的次数，通常 k≤n 且 t≤n.  

采用一种较为科学的方法确定合理的k值（绘制碎石图分析）  
sumsq<-NULL  
for (i in 1:15) sumsq[i] <- sum(kmeans(newthree,centers=i,iter.max=5000)$withinss)  
plot(1:15,sumsq,type="b",xlab="Number of Clusters",ylab="Within groups sum of squares",main="screeplot")  


我们将k值从1取到15，迭代15次，得到此碎石图，横坐标为k，纵坐标为组内平方和。理想的聚类数为碎石图拐点（陡坡和缓坡的临界点）所在的位置，即k=3。  

预测新数据的聚类从属关系  
#给数据集添加聚类信息  
Cluster <- cbind(newthree,Membership=knewthree\$cluster)  
#预测新数据的聚类从属关系  
predict.kmeans <- function(km,newdata){  
	k <- nrow(km\$centers)  
	d <- as.matrix(dist(rbind(km\$centers,newdata)))[-(1:k),1:k]  
	n <- nrow(newdata)  
	out <- apply(d,1,which.min)  
	return(out)  
}  
#使用table进行统计分析  
Cluster\$Predicted <- predict.kmeans(knewthree,newthree)  
table(Cluster\$Membership,Cluster\$Predicted)  


从图中使用了table进行统计分析，每一行代表原始数据的结果，每一列代表对其进行预测得到的结果，我们自己编写算法预测聚类信息要原始数据是完全一致的。
